```python
import pandas as pd
from IPython.display import display
```

```python
df_dict = {'i' : [],
           'j':[],
           'k':[]}
df_dict
```

```python
for i in range(0,11): 
    for j in range(0,11):
        for k in range(0,11):
            df_dict['i'] += [i]
            df_dict['j'] += [j]
            df_dict['k'] += [k]
df = pd.DataFrame(df_dict)
display(df)
df.shape
```

```python
df['total'] = df['i'] + df['j'] + df['k']
display(df)
```

```python
df2 = df[df['total'] == 10].reset_index(drop = True)
display(df2)
```
